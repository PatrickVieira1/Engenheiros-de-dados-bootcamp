## Atividades
1. Vamos criar uma classe para isso. Comece criando um método para
abrir os arquivos e armazená-los como atributos da classe. Para isso,
abra os arquivos 'worldwide internet users - users.csv', 'worldwide
internet prices in 2022 - IN 2022.csv' e 'worldwide internet speed in
2022 - avg speed.csv' usando PANDAS e os deixe armazenados em
atributos: usuarios, precos e velocidades, respectivamente

```python
class Database():
    def __init__(self, usuarios=pd.read_csv(r"datasets\users.csv"), precos=pd.read_csv(r"datasets\prices.csv"), velocidades=pd.read_csv(r'''datasets\avgspeed.csv''')):
        self.usuarios = usuarios
        self.precos = precos
        self.velocidades = velocidades
        
```

```python
Database = Database()
```

```python
def unificador_dados(self):
    dados_unificados = self.usuarios.set_index('Country or area').join(self.precos.set_index('Name'), how='outer').join(self.velocidades.set_index('Country'), how='outer')
    return dados_unificados

unificador_dados(Database)
```

```python
dados_sem_character = limpeza_população(Database).replace('\$','', regex=True)

dados_sem_character.head()
```

```python
dados_sem_character['Average price of 1GB (USD)'] = pd.to_numeric(dados_sem_character['Average price of 1GB (USD)'], errors='coerce')
dados_sem_character = dados_sem_character[dados_sem_character['Average price of 1GB (USD)'].notnull()]

dados_sem_character.head()
```

```python
dados_sem_character['Average price of 1GB (USD)'].median()
dados_sem_character['NO. OF Internet Plans '].median()
```

```python
dados_sem_character['Cheapest 1GB for 30 days (USD)'] = pd.to_numeric(dados_sem_character['Cheapest 1GB for 30 days (USD)'], errors='coerce')
dados_sem_character = dados_sem_character[dados_sem_character['Cheapest 1GB for 30 days (USD)'].notnull()]
```

```python
dados_sem_character['Most expensive 1GB (USD)'] = pd.to_numeric(dados_sem_character['Most expensive 1GB (USD)'], errors='coerce')
dados_sem_character = dados_sem_character[dados_sem_character['Most expensive 1GB (USD)'].notnull()]
```

```python
dados_sem_character['Average price of 1GB (USD  at the start of 2021)'] = pd.to_numeric(dados_sem_character['Average price of 1GB (USD  at the start of 2021)'], errors='coerce')
dados_sem_character = dados_sem_character[dados_sem_character['Average price of 1GB (USD  at the start of 2021)'].notnull()]
```

```python
dados_sem_character['Average price of 1GB (USD – at start of 2020)'] = pd.to_numeric(dados_sem_character['Average price of 1GB (USD – at start of 2020)'], errors='coerce')
dados_sem_character = dados_sem_character[dados_sem_character['Average price of 1GB (USD – at start of 2020)'].notnull()]
```

```python
dados_sem_character['AnaliseQuantidadePlanos'] = np.where(dados_sem_character['NO. OF Internet Plans '] > dados_sem_character['NO. OF Internet Plans '].median(), 'MUITO', 'POUCO')
```

```python
dados_sem_character['AnalisePrecoMedio'] = np.where(dados_sem_character['Average price of 1GB (USD)'] >= dados_sem_character['Average price of 1GB (USD)'].median(), 'CARO', 'BARATO')
```

```python
dados_sem_character['AnalisePrecoMedio'].value_counts()
```

```python
dados_sem_character['AnaliseQuantidadePlanos'].value_counts()
```

```python
dados_sem_character['PerfilPlanoPreco'] = dados_sem_character['AnalisePrecoMedio'] + dados_sem_character['AnaliseQuantidadePlanos']	

dados_sem_character.head()
```
