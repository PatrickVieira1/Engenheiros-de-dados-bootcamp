## Atividades
1. Vamos criar uma classe para isso. Comece criando um método para
abrir os arquivos e armazená-los como atributos da classe. Para isso,
abra os arquivos 'worldwide internet users - users.csv', 'worldwide
internet prices in 2022 - IN 2022.csv' e 'worldwide internet speed in
2022 - avg speed.csv' usando PANDAS e os deixe armazenados em
atributos: usuarios, precos e velocidades, respectivamente

```python
class Database():
    def __init__(self, usuarios=pd.read_csv('datasets\worldwide internet users - users.csv'), precos=pd.read_csv('datasets\worldwide internet prices in 2022 - IN 2022.csv'), velocidades=pd.read_csv('datasets\worldwide internet speed in 2022  - avg speed.csv')):
        self.usuarios = usuarios
        self.precos = precos
        self.velocidades = velocidades
        
```

```python
Database = Database()
```

```python
def unifica_dados(self):
    df = pd.merge(self.precos, self.usuarios, left_on='Name', right_on='Country or area')
    self.dados_unificados = pd.merge(df, self.velocidades, left_on='Country or area', right_on='Country')
    del self.dados_unificados['Name']
    return self.dados_unificados
```

```python
def limpeza_dados(self):
    # self.dados_unificados['Average price of 1GB (USD)'] = self.dados_unificados['Average price of 1GB (USD)'].str.replace('$', '')
    # self.dados_unificados['Cheapest 1GB for 30 days (USD)'] = self.dados_unificados['Cheapest 1GB for 30 days (USD)'].str.replace('$', '')
    # self.dados_unificados['Most expensive 1GB (USD)'] = self.dados_unificados['Most expensive 1GB (USD)'].str.replace('$', '')
    # self.dados_unificados['Average price of 1GB (USD  at the start of 2021)'] = self.dados_unificados['Average price of 1GB (USD  at the start of 2021)'].str.replace('$', '')
    # self.dados_unificados['Average price of 1GB (USD – at start of 2020)'] = self.dados_unificados['Average price of 1GB (USD – at start of 2020)'].str.replace('$', '')
    self.dados_unificados['Average price of 1GB (USD)'] = pd.to_numeric(self.dados_unificados['Average price of 1GB (USD)'], errors='coerce').astype(float)
    self.dados_unificados['Cheapest 1GB for 30 days (USD)'] = pd.to_numeric(self.dados_unificados['Cheapest 1GB for 30 days (USD)'], errors='coerce', downcast='float').astype(float)
    self.dados_unificados['Most expensive 1GB (USD)'] = pd.to_numeric(self.dados_unificados['Most expensive 1GB (USD)'], errors='coerce', downcast='float').astype(float)
    self.dados_unificados['Average price of 1GB (USD  at the start of 2021)'] = pd.to_numeric(self.dados_unificados['Average price of 1GB (USD  at the start of 2021)'], errors='coerce', downcast='float').astype(float)
    self.dados_unificados['Average price of 1GB (USD – at start of 2020)'] = pd.to_numeric(self.dados_unificados['Average price of 1GB (USD – at start of 2020)'], errors='coerce', downcast='float').astype(float)
    self.dados_unificados = self.dados_unificados.dropna(axis=0, how='any')
    return self.dados_unificados
```
